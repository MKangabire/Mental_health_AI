## Tokenization and Normalization
- **Tokenization**: We use the `BlenderbotSmallTokenizer`, which implements Byte-Pair Encoding (BPE). BPE splits text into subword units, starting with individual characters and iteratively merging frequent pairs to form a vocabulary. This approach is effective for handling rare or out-of-vocabulary words, making it suitable for conversational datasets like mental health counseling, where varied expressions may occur.
- **Why BPE?**: Compared to WordPiece (used in BERT), BPE is simpler and does not rely on a predefined vocabulary size. It’s well-suited for Blenderbot, which is designed for open-domain dialogue. BPE ensures robust handling of diverse user inputs.
- **Normalization**: The tokenizer handles basic normalization (e.g., Unicode normalization). We additionally clean the text by removing URLs, special characters, and extra whitespace to standardize inputs. Lowercasing is not applied, as the tokenizer is case-sensitive and preserves meaning in conversational text.
- **Parameters**: We set `max_length=128` to balance context retention and computational efficiency, with `truncation=True` to handle longer texts and `padding="max_length"` to ensure uniform input sizes.

```python
tokenizer = BlenderbotSmallTokenizer.from_pretrained("facebook/blenderbot_small-90M")
model = BlenderbotSmallForConditionalGeneration.from_pretrained("facebook/blenderbot_small-90M")
```

```python
tokenizer = BlenderbotSmallTokenizer.from_pretrained("facebook/blenderbot_small-90M")
model = BlenderbotSmallForConditionalGeneration.from_pretrained("facebook/blenderbot_small-90M")
def preprocess_function(examples):
    inputs = ["User: " + text for text in examples['Context']]  # Add context prefix
    targets = ["Bot: " + text for text in examples['Response']]  # Add response prefix
    model_inputs = tokenizer(inputs, max_length=128, truncation=True, padding="max_length")
    labels = tokenizer(targets, max_length=128, truncation=True, padding="max_length")
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

tokenized_dataset = dataset.map(preprocess_function, batched=True)
```

```python
# Define training arguments
training_args = TrainingArguments(
    output_dir="./fine_tuned_blenderbot",
    num_train_epochs=20,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=16,
    gradient_accumulation_steps=2,
    lr_scheduler_type="cosine",
    save_strategy="epoch",
    logging_dir="./logs",
    logging_steps=100,
    learning_rate=5e-5,
    report_to="none",  # Disable wandb logging
    eval_strategy="steps",
    warmup_steps =500,
    weight_decay = 0.05
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset['train'],
    eval_dataset=tokenized_dataset['test'],
)

# Fine-tune the model
trainer.train()
```

```python
# Evaluation function
def calculate_metrics(dataset, model, tokenizer):
    bleu_evaluator = load("bleu")
    rouge_evaluator = load("rouge")
    predictions, references = [], []
    perplexities = []

    model.eval()
    for example in dataset['test']:
        input_text = example['Context']
        reference = example['Response']

        # Generate prediction
        inputs = tokenizer(input_text, return_tensors="pt", max_length=128, truncation=True, padding=True)
        inputs = {key: val.to(device) for key, val in inputs.items()}
        outputs = model.generate(
            input_ids=inputs["input_ids"],
            attention_mask=inputs["attention_mask"],
            max_length=128,
            num_beams=5,
            no_repeat_ngram_size=2,
            top_k=50,
            top_p=0.95,
            do_sample=True,
        )
        prediction = tokenizer.decode(outputs[0], skip_special_tokens=True)
        predictions.append(prediction)
        references.append(reference)

        # Calculate Perplexity
        actual_sequence = input_text + " " + reference
        encodings = tokenizer(actual_sequence, return_tensors="pt", max_length=128, truncation=True, padding=True)
        encodings = {key: val.to(device) for key, val in encodings.items()}
        with torch.no_grad():
            outputs = model(**encodings, labels=encodings["input_ids"])
            loss = outputs.loss
            perplexity = math.exp(loss.item())
        perplexities.append(perplexity)

    # Compute BLEU and ROUGE
    bleu_results = bleu_evaluator.compute(predictions=predictions, references=references)
    rouge_results = rouge_evaluator.compute(predictions=predictions, references=references)

    # Return raw results
    return {
        "BLEU": bleu_results,
        "ROUGE": rouge_results,
        "Perplexity": perplexities
    }

# Evaluate metrics
metrics = calculate_metrics(dataset, model, tokenizer)
print("Evaluation Metrics:")
print("BLEU:", metrics["BLEU"])
print("ROUGE:", metrics["ROUGE"])
print("Perplexity:", metrics["Perplexity"])
```
